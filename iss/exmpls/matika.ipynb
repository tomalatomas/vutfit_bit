```python
# -------------- Matematika pro ISS --------------------------
import numpy as np
import matplotlib.pyplot as plt

# --- soucet signalu ---
t = np.arange(-5,5,0.01)
x1 = np.zeros(t.shape)
ii = np.argwhere((t>-1) & (t<0.5)) 
x1[ii] = 0.5
x2 = np.zeros(t.shape)
ii = np.argwhere((t>-0.5) & (t<2)) 
x2[ii] = 1
y = x1 + x2
_, ax = plt.subplots(3,1, figsize=(10,5))
ax[0].plot(t,x1)
ax[0].set_ylabel('$x_1(t)$')
ax[0].set_ylim(-0.1,2)
ax[0].grid()
ax[1].plot(t,x2,color='g')
ax[1].set_ylabel('$x_2(t)$')
ax[1].set_ylim(-0.1,2)
ax[1].grid()
ax[2].plot(t,y,color='r',linewidth=3)
ax[2].set_ylabel('$y(t)$')
ax[2].set_xlabel('$t$')
ax[2].set_ylim(-0.1,2)
ax[2].grid()
plt.savefig('FIG/soucet.pdf')
```

```python
# ----------- linearni funkce --------------
t = np.arange(-5,5,0.01)
x1 = 3 * t; 
x2 = 2*t + 1; 
x3 = -t; 
x4 = -2*t -2; 
plt.figure(figsize=(5,4))
plt.plot(t,x1,label="$x_1(t) = 3t$")
plt.plot(t,x2,label="$x_2(t) = 2t+1$") 
plt.plot(t,x3,label="$x_3(t) = -t$") 
plt.plot(t,x4,label="$x_4(t) = -2t-2$") 
plt.gca().set_xlabel('$t$')
plt.gca().grid()
plt.legend()
plt.savefig('FIG/lin.pdf')
```

```python
# ---------- derivace, primitivni funkce a integral ------------------
t = np.arange(-3,2,0.05)
x = np.power(t,2) + 2*t + 3 
xd = 2*t + 2  # derivace
xp = np.power(t,3) / 3 + np.power(t,2) + 3*t # prim. funkce
plt.figure(figsize=(8,6))
plt.plot(t,x,label="$x(t) = t^2 + 2t + 3$")
plt.plot(t,xd,label="derivace") 
plt.plot(t,xp,label="prim. funkce") 
plt.gca().set_xlabel('$t$')
plt.gca().grid()
plt.legend()
plt.savefig('FIG/polynom_derivace_prim.pdf')
```
